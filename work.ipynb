```python
import pandas as pd

def extract_video_name_list(onlypos_labels_path, basename, save=True):
    base_video_name_list = pd.read_csv('/mnt/iot-qnap3/mochida/medical-care/InitialModelConstruction/data/labels/PIMD_A/comfort-ausign-video_name_list-fbf.csv')
    onlypos_labels = pd.read_csv(onlypos_labels_path)
    onlypos_video_list = onlypos_labels['video_name'].unique()
    new_video_name_list = pd.DataFrame()
    
    for i in range(len(base_video_name_list)):
        base_video_list = list(base_video_name_list.iloc[i, :][base_video_name_list.iloc[i, :] == 1].index)
        
        for video_name in onlypos_video_list:
            if video_name in base_video_list:
                new_video_name_list = pd.concat([new_video_name_list, base_video_name_list.iloc[i, :]], axis=1)
                break
    
    new_video_name_list = new_video_name_list.T
    if save:
        new_video_name_list.to_csv('/mnt/iot-qnap3/mochida/medical-care/InitialModelConstruction/data/labels/PIMD_A/comfort-ausign-video_name_list-fbf-{}.csv'.format(basename), index=False)
    
    return new_video_name_list
```
